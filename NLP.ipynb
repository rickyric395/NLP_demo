# Creating word of bag

```python
# removing all non- alphanetic , non-numeric character
import re
dataset = nltk.sent_tokenize(para)
for i in range(len(dataset)):
    dataset[i] = dataset[i].lower()
    dataset[i] = re.sub(r'\W',' ',dataset[i])
    dataset[i] = re.sub(r'\s+',' ',dataset[i])
```

```python
dataset
```

```python
#creating a histogram for bag of words
word_count = {}
for data in dataset:
    words = nltk.word_tokenize(data)
    for word in words:
        if word not in word_count.keys():
            word_count[word] = 1
        else:
            word_count[word] += 1
```

```python
word_count
```

```python
# creating the list of most frequent words in wordcount
import heapq
freq_count = heapq.nlargest(100,word_count,word_count.get)
freq_count
```

```python
# creating a vector of size 0 and 1
X = []
import numpy as np
for data in dataset:
    vector = []
    for word in freq_count:
        if word in nltk.word_tokenize(data):
            vector.append(1)
        else:
            vector.append(0)
    X.append(vector)
X = np.asarray(X)
X
```

# TF-IDF Better over bag of words


```python
import re
import numpy as np
dataset1 = nltk.sent_tokenize(para)
for i in range(len(dataset1)):
    dataset1[i] = dataset1[i].lower()
    dataset1[i] = re.sub(r'\W',' ',dataset1[i])
    dataset1[i] = re.sub(r'\s+',' ',dataset1[i])
```

```python
word_freq1 = {}
for data in dataset1:
    words = nltk.word_tokenize(data)
    for word in words:
        if word not in word_freq1:
            word_freq1[word] = 1
        else:
            word_freq1[word] += 1
word_freq1
```

```python
import heapq
freq_count = heapq.nlargest(100,word_freq1,key = word_freq1.get)
```

```python
# calculating idf value
word_idf = {}
for words in word_freq1:
    doc_count = 0
    for data in dataset1:
        if words in nltk.word_tokenize(data):
            doc_count += 1
    word_idf[words] = np.log((len(dataset1)/doc_count)+1)
word_idf
```

```python
#calculating tf value
tf_values = {}
for word in freq_count:
    freq = 0
    tf_list = []
    for data in dataset1:
        for w in nltk.word_tokenize(data):
            if w == word:
                freq += 1
        tf_list.append(freq/len(nltk.word_tokenize(data)))
    tf_values[word] = tf_list
```

```python
# calculating tf-idf values
tf_idf = []
for word in tf_values.keys():
    temp_tf_idf = []
    for value in tf_values[word]:
        temp_tf_idf.append(word_idf[word] * value)
    tf_idf.append(temp_tf_idf)
        
```

```python
tf_idf = np.asarray(tf_idf)
tf_idf.shape
```

```python
X = np.transpose(tf_idf)
X.shape
```

# Building N-Gram Model
### Character N-Gram Model

```python
import random

sen = """Global warming or climate change has become a worldwide concern. It is gradually developing into an unprecedented environmental crisis evident in melting glaciers, changing weather patterns, rising sea levels, floods, cyclones and droughts. Global warming implies an increase in the average temperature of the Earth due to entrapment of greenhouse gases in the earth’s atmosphere."""
```

```python
#Character n gram ( trigram )
n = 3 #n of n-gram
ngrams = {}
for i in range(len(sen)-n):
    gram = sen[i:i+n]
    if gram not in ngrams.keys():
        ngrams[gram] = []
    ngrams[gram].append(sen[i+n])
```

```python
#testing trigram
currentgram = sen[0:n]
result = currentgram
for i in range(100):
    if currentgram not in ngrams.keys():
        break
    possible = ngrams[currentgram]
    next_item = possible[random.randrange(len(possible))]
    result += next_item
    currentgram = result[len(result)-n:len(result)]
```

# Word Negation Tracking 

```python
sen = 'I am not not good at parenting'
```

```python
words = nltk.word_tokenize(sen)
new_words = []
temp = ''
for word in words:
    antonyms = []
    if word == 'not':
        temp = 'not_'
    elif temp == 'not_':
        for syn in wordnet.synsets(word):
            for s in syn.lemmas():
                for a in s.antonyms():
                    antonyms.append(a.name())
        if len(antonyms) >= 1:
            word = antonyms[0]
        else:
            word = temp + word
        temp = ''
    if word != 'not':
        new_words.append(word)
```
